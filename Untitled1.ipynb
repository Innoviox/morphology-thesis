```python
import collections
def check(inputs, output, model_output):
  num_correct_i, num_total_i = 0, 0
  num_c_d, num_t_d = 0, 0
  res = [collections.defaultdict(list) for _ in range(6)]
  for ii, a, b in zip(inputs, output, model_output):
    kk = ii.split()
    corr = 0
    # q = a
    a = a.strip("- ").split()# [6:]
    b = b.strip("- ").split()# [6:]
    if ii.split()[0].isupper():
      num_t_d += 1
      if all(c == d for [c, d] in zip(a, b)):
        # print(ii, q, a, b)
        # input()
        corr = 1
        num_c_d += 1
    else:
      num_total_i += 1
      if all(c == d for [c, d] in zip(a, b)):
        corr = 1
        num_correct_i += 1
    for j in range(6):
      res[j][kk[j]].append(corr)
  for i, d in enumerate(res):
    print(i)
    for k, v in d.items():
      if "," in k: continue
      print("\t", k, len(v), sum(v) / len(v))
  print(num_correct_i / num_total_i, num_c_d / num_t_d)
  return (num_correct_i + num_c_d) / (num_total_i + num_t_d)
  return (num_correct_i / num_total_i), (num_c_d / num_t_d)

inputs = open("test_in.txt").readlines()
output = open("test_out.txt").readlines()
```

```python
test = open("regbP/test_decoded_lesioned_0.txt").readlines()
print(check(inputs, output, test))
```

```python
# accuracies = {}
points = []
colors = []

regbP_output = open("regbP/test_decoded_lesioned_0.txt").readlines()
i, d = check(inputs, output, regbP_output)
points.append([i, d])
colors.append(0)

for lesion in range(5, 100, 5):
    # accuracies[lesion] = [0, 0]
    for run in 'abcde':
        f = open(f"regbP/test_decoded_lesioned_{lesion}{run}.txt").readlines()
        # accuracies[lesion].append(check(inputs, output, f))
        i, d = check(inputs, output, f)
        points.append([i, d])
        colors.append(lesion / 100)

    # accuracies[lesion] = sum(accuracies[lesion]) / len(accuracies[lesion])
# print(points)
```
